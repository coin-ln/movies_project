```python
import pandas as pd
```

```python
data = pd.read_excel(r'E:\graduate\movies_project\Models\国外的玩意\data2.xlsx')
data0 = pd.read_excel((r'E:\graduate\movies_project\Models\国外的玩意\data1.xlsx'))
data.head
```

```python
data = data.iloc[:122,:9]
data0 = data0.iloc[:122,:9]
data
```

```python
duration  = data['片长']
duration0 = data0['片长']
duration
```

```python
duration = [int(item[:-3]) for item in duration]
duration0 = [int(item[:-3]) for item in duration0]

duration
```

```python
data['片长'] = duration
data0['片长'] = duration0
print(data)
```

```python
data = pd.concat([data,data0],axis=0,ignore_index=True)
def find_earliest_year(group):
    # 对分组内的数据按年份进行排序
    sorted_group = group.sort_values('年份')
    # 取年份最小的行
    earliest_year_row = sorted_group.iloc[0]
    # 累计票房
    earliest_year_row['年度票房'] = sorted_group['年度票房'].sum()
    return earliest_year_row

# 使用groupby和apply函数应用上面的函数
data_result = data.groupby('电影名称').apply(find_earliest_year).reset_index(drop=True)
data_result
```

```python
data_result.to_csv('data_worked.csv')
```
